## 2. Model

A small CNN — deliberately simple. The point of this project is showing you understand the pipeline end-to-end, not building a state-of-the-art architecture.

```python
class SimpleMaterialCNN(nn.Module):
    def __init__(self, num_classes: int):
        super().__init__()
        self.features = nn.Sequential(
            nn.Conv2d(3, 16, kernel_size=3, padding=1),
            nn.ReLU(),
            nn.MaxPool2d(2),  # 64x64 -> 32x32
            nn.Conv2d(16, 32, kernel_size=3, padding=1),
            nn.ReLU(),
            nn.MaxPool2d(2),  # 32x32 -> 16x16
        )
        self.classifier = nn.Sequential(
            nn.Flatten(),
            nn.Linear(32 * 16 * 16, 128),
            nn.ReLU(),
            nn.Linear(128, num_classes),
        )

    def forward(self, x):
        x = self.features(x)
        x = self.classifier(x)
        return x

# TODO: set num_classes to match your actual dataset
# model = SimpleMaterialCNN(num_classes=4).to(device)
print("Model class defined - instantiate once num_classes is known from your dataset")
```

## 3. Training Loop

```python
def train_one_epoch(model, loader, optimizer, criterion):
    model.train()
    total_loss = 0.0
    for images, labels in loader:
        images, labels = images.to(device), labels.to(device)
        optimizer.zero_grad()
        outputs = model(images)
        loss = criterion(outputs, labels)
        loss.backward()
        optimizer.step()
        total_loss += loss.item()
    return total_loss / len(loader)

# TODO once dataset + model are set up:
# optimizer = optim.Adam(model.parameters(), lr=1e-3)
# criterion = nn.CrossEntropyLoss()
# for epoch in range(10):
#     avg_loss = train_one_epoch(model, train_loader, optimizer, criterion)
#     print(f"Epoch {epoch+1}: loss={avg_loss:.4f}")
```
